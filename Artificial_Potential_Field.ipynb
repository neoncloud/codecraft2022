```python
import numpy as np
import math

# 障碍物的坐标
obstacles = [(1,1), (2,2), (3,3)]
# 机器人当前位置坐标
current_coord = np.array([2, 1])
# 机器人当前速度矢量
current_vel = np.array([1, 1])
# 机器人当前角速度
current_w = -0.3
# 目的地坐标
goal_coord = np.array([10, 10])
# 势场常数
k_att = 0.25 # 表示机器人与目标之间的引力常数
k_rep = 2.0 # 表示机器人与障碍物之间的斥力常数
obs_rep_range = 2.0 # 表示障碍物斥力的作用范围
bound_rep_range = 0.1 # 表示边界斥力的作用范围
KW = 2 # 超参数
KV = 2 # 超参数
# 地图大小
map_size = np.array([50, 50])
def potential_field(current_pos, goal_pos):
    """
    计算机器人在当前位置的势场
    """
    att = k_att * (goal_pos - current_pos) # 计算机器人当前位置和目标位置之间的引力
    obs_rep = np.float64(np.array([0, 0])) # 用于计算机器人当前位置与障碍物之间的斥力
    for obs in obstacles:
        obs_vec = np.float64(current_pos - np.array(obs))
        d = np.float64(np.linalg.norm(obs_vec))
        if d < obs_rep_range:
            obs_rep += obs_rep * (1/d - 1/obs_rep_range) * (1/d**2) * obs_vec / d
    # 添加地图边界限制条件
    boundary_rep = np.float64(np.array([0, 0]))
    if current_pos[0] < bound_rep_range:
        boundary_rep[0] = k_rep * (1/(current_pos[0]+1e-6) - 1/bound_rep_range) / (current_pos[0]+1e-6)**2
    elif current_pos[0] > map_size[0] - bound_rep_range:
        boundary_rep[0] = -k_rep * (1/(map_size[0]-current_pos[0]+1e-6) - 1/bound_rep_range) / (map_size[0]-current_pos[0]+1e-6)**2
    if current_pos[1] < bound_rep_range:
        boundary_rep[1] = k_rep * (1/(current_pos[1]+1e-6) - 1/bound_rep_range) / (current_pos[1]+1e-6)**2
    elif current_pos[1] > map_size[1] - bound_rep_range:
        boundary_rep[1] = -k_rep * (1/(map_size[1]-current_pos[1]+1e-6) - 1/bound_rep_range) / (map_size[1]-current_pos[1]+1e-6)**2
    # print('att = ', att, 'rep = ', rep, 'boundary_rep = ', boundary_rep)
    return att + obs_rep + boundary_rep
# 计算机器人需要更改的速度矢量
# new_vel = current_vel + potential_field(current_coord, goal_coord)
new_vel = potential_field(current_coord, goal_coord)
# 输出结果
print("机器人原来的速度矢量为：", current_vel)
print("机器人需要更改的速度矢量为：", new_vel)

angle_needed = np.arctan2(np.cross(current_vel, new_vel), np.dot(current_vel, new_vel))
print('angle_needed=',angle_needed) # 输出角度制的夹角
w = KW*(angle_needed - current_w)
v = KV*np.linalg.norm(new_vel)
print('current_w=',current_w)
print('w=',w)
print('v=',v)
```
